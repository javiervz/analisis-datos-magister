## 1. pre-procesamiento de los datos ##

```python
# Para iniciar, se descargan todos los archivos csv individualizados para la variación de cada palabra en SC. 
# Luego de tener todos los archivos en un directorio común,
# la idea es juntarlos en un solo dataframe. 
```

```python
import pandas as pd

import glob

import os
```

```python
# Luego, con la función "glob", se genera una lista con todos los archivos csv reunidos en dicho directorio. 
```

```python
all_csv = glob.glob("- copiar directorio -/Datos/*.csv")
```

```python
# Seguidamente, se debe crear una lista donde se pondrán todas las lecturas de datos en la lista "all_csv".
```

```python
csv_list = [] # lista vacía #
```

```python
df = pd.concat(csv_list, ignore_index = True) # se crea el dataframe #
```

## 2. Técnica de análisis de datos ##

```python
# Luego del pre-procesamiento de datos, se implementa la función que permite calcular la distancia de Levenshtein.
```

```python
def funcion_distancia(string1,string2): # la función recibe dos strings diferentes #
    m = len(string1) # m es el tamaño del string 1 #
    n = len(string2) # n es el tamaño del string 2 #
    table = [[0] * (n+1) for _ in range (m+1)] # tabla con el número de tablas de m+1 y n+1 elementos en ellas #

    for i in range(m+1): # recorre cada caracter en el rango de 0 a m #
        table[i][0] = i # lista con las listas de caracteres de m, i es el tamaño del string #
    for j in range(n+1): # recorre cada caracter en el rango de 0 a n #
        table[0][j] = j # lista con las listas de caracteres de n, j es el tamaño del string #
        
    for i in range(1, m+1): # recorre cada elemento i para el rango entre 1 y el tamaño del string 1 #
        for j in range(1, n+1): # recorre cada elemento j para el rango entre 1 y el tamaño del string 2 #
            if string1[i - 1] == string2[j - 1]: # si el string 1 es igual al string 2 ... #
                table[i][j] = table[i - 1][j - 1] # lista para los rangos de cada string #
            else:
                table[i][j] = 1 + min(table[i - 1][j], table[i][j - 1], table[i - 1][j - 1]) # si no, ... 

    return table [-1][-1] # devuelve la medida para la diferencia entre las listas #
```

```python
# Se prueba la función con un ejercicio: medir la distancia entre dos pronunciaciones de la palabra "adümchefe":
# la primera pronunciación corresponde a la localidad Santa Bárbara y la segunda a Alto Bio-Bío.
```

```python
funcion_distancia('ɐ̝.ˌðʊm.ˈt͡ʃë.vë','ˌɐ̝.ðʊm.ˈt͡ʃë.ʋë')
```

```python
# la distancia entre ambas pronunciaciones es 3.
```

```python
# luego de probada la función, la idea es construir dos diccionarios utilizando Pandas:
# 1. uno que contenga las palabras seleccionadas a medir,
# 2. y otro que acumule los datos de las mediciones.
```

```python
# lo primero es crear dos set de datos desde el dataframe
```

```python
names = set(dataframe['LanguageName']) # selecciona un set de datos de las localidades del dataframe #
```

```python
palabras = set(dataframe['WordModernName1']) # selecciona las palabras disponibles del dataframe #
```

```python
palabras_elegidas = ['persona_que_ensenya'] # lista con los datos de una de las palabras 'adümchefe' (profesor) #
```

```python
dict_dataframe = {L:{} for L in names} # diccionario con los datos de localidades en Key y las variaciones en Values #
```

```python
for L in names: # recorre todas las localidades disponibles en el diccionario 'names' #
    for word in palabras_elegidas: # recorre todas las variaciones de la palabra escogida #
        D = dataframe[dataframe['LanguageName']==L] # D es un dataframe que selecciona a todas las localidades #
        try:
            dict_dataframe[L][word]=list(D[D['WordModernName1']==word]['Phonetic'])[0] # agrega al diccionario, por cada localidad en Key, palabra escogida y su pronunciación en Values # 
        except IndexError:
            dict_dataframe[L][word]='unk' # dejar como 'unk' si no hay información para la localidad consultada #
```

```python
dict_dataframe
```

```python
import itertools # iteradores para bucles #
```

```python
# se elabora una lista iterando y contraponiendo las localidades existentes en dict_dataframe.
```

```python
pares_localidades = list(itertools.product(list(dict_dataframe.keys()), list(dict_dataframe.keys())))
```

```python
pares_localidades
```

```python
# Ahora bien, antes de calcular las distancias entre todos estos pares,
# es necesario crear un diccionario que acumule todas las mediciones.
```

```python
distancias = {L:{LL:0 for LL in list(dict_dataframe.keys())} for L in list(dict_dataframe.keys())}
```

```python
distancias # como aun no se aplica la función de Levenshtein al dict_dataframe, todas las localidades tienen valor 0 #
```

```python
# Ahora se realiza un ciclo for para aplicar 'funcion_distancia' a la organización de las variables.
```

```python
for par in pares_localidades: # se recorren cada par de palabras en lista "pares_localidades"#
    distancias[par[0]][par[1]] = funcion_distancia(dict_dataframe[par[0]]['persona_que_ensenya'],dict_dataframe[par[1]]['persona_que_ensenya'])
```

```python
distancias
```

```python
# como la palabra "adümchefe" se encuentra solo en 15 de las 37 localidades disponibles,
# la mayoría de los valores se encuentran en 16, 17, 18 o 19, debido a que en la medición de pares
# el values que no tiene la información se marca como 'unk' y el cálculo se efectúa entre 3 y la distancia del string.
```

```python
# En definitiva, se toman todos los valores del diccionario 'distancias' menores a 10, 
# los que indican que se hizo efectivamente el cálculo de distancia.
```
